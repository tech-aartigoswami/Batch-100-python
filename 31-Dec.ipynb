### Data Types in python
- string str()
- number(integer int(),float,float())
- boolean(True(),False())

```python
# 1. string str()
str1 = 'Dev'
str2 = "python 'and' programming"
# multiline string
str3 = """python 
'24'
programming"""
str4 = '''python programming'''
print(str1)
print(str2)
print(str3)
print(str4)
str5 = f"""this is python file {str1}"""
print(type(str1))
print(str5)
```

```python
str5 = f"""this is python file {str1}"""
print(str5)
```

```python
str6 = "this" "is" "python"
print(len(str6))
print(str6)
```

```python
# 2. number(integer int(),float,float())
num1 = 56
print(num1)
print(type(num1))

num2 = 32.5
print(num2)
print(type(num2))
```

```python
# 3. boolean(True,False)
b1 = True
b2 = False
print(type(b1))
print(b1+b2)

```
